```python
from gensim.models import Word2Vec

# 훈련에 사용될 데이터 (리스트의 리스트 형태로 변경)
training_data = [
    ['강아지', '고양이', '두', '마리', '계단', '위', '앉아', '있다']
]

# Word2Vec 사용하여 벡터로 변환
word2vec_model = Word2Vec(sentences=training_data, min_count=1)

# '강아지' 단어의 벡터 표현
word_vector = word2vec_model.wv['강아지']
word_vector
```
